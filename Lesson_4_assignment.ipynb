**5. Concatenation and Merging**

```python
# DataFrames
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Kodi', 'Keilah']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name': ['Kacee', 'Shanon']})
df3 = pd.DataFrame({'ID': [1, 2, 3, 4], 'Score': [85, 90, 95, 88]})

# Concatenate df1 and df2 vertically
df_concat = pd.concat([df1, df2], ignore_index=True)
df_concat
```

**9. Iterators and Generators**

```python
# Iterator class that generates first 5 even numbers
class EvenIterator:
    def __init__(self):
        self.n = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= 5:
            raise StopIteration
        result = self.n
        self.n += 2
        self.count += 1
        return result

# Generator function that generates first 5 even numbers
def even_generator():
    n = 0
    for _ in range(5):
        yield n
        n += 2

print("Iterator output:", list(EvenIterator()))
print("Generator output:", list(even_generator()))
```

**10. Maps, Reduce, Filters**

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map
squared = list(map(lambda x: x**2, nums))
print("Squared using map:", squared)

# Reduce
product = reduce(lambda x, y: x * y, nums)
print("Product using reduce:", product)

# Filter
evens = list(filter(lambda x: x % 2 == 0, nums))
print("Even numbers using filter:", evens)
```

**11. Object-Oriented Programming**

```python
# Rectangle class with attributes for length and width
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create 2 rectangular objects
rect1 = Rectangle(7, 3)
rect2 = Rectangle(10, 4)

print("Rect1_Area:", rect1.area(), ",Perimeter:", rect1.perimeter())
print("Rect2_Area:", rect2.area(), ",Perimeter:", rect2.perimeter())
```
